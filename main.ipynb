## Bayesian Ridge Regression for disaggreagtion of missing district counts

```python
## df2 prediction

from sklearn.linear_model import BayesianRidge
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
# Filter out merged districts in 2021 based on lookup mismatches
merged_codes = lookup[lookup['LAD11CD'] != lookup['LAD21CD']]['LAD21CD'].unique()
df_2021_merged = df2[df2['Area Code'].isin(merged_codes)]

# Get disaggregated components from 2011 for those merged codes
merged_map = lookup[lookup['LAD21CD'].isin(merged_codes)]
disaggregated_2011 = df1[df1['Area Code'].isin(merged_map['LAD11CD'])]

# Train Bayesian Ridge Regression on 2011 data
features = ['Tenure', 'Accomodation Type']
target = 'Counts'

preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), features)
])

model = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', BayesianRidge())
])

X_train = disaggregated_2011[features]
y_train = disaggregated_2011[target]
model.fit(X_train, y_train)

# Prepare disaggregated 2021 structure by duplicating pre-merger rows and assigning them under the merged LAD
df_split_list = []

for merged_code in merged_codes:
    # Get merged row(s)
    merged_rows = df_2021_merged[df_2021_merged['Area Code'] == merged_code]

    # Get original LAD11CDs contributing to this LAD21CD
    original_parts = merged_map[merged_map['LAD21CD'] == merged_code]

    for _, merged_row in merged_rows.iterrows():
        # Build test set: duplicate merged_row for each original LAD
        temp = pd.DataFrame([merged_row] * len(original_parts))
        temp['Area Code'] = original_parts['LAD11CD'].values
        temp['Area Names'] = original_parts['LAD11NM'].values
        temp['Merged District Code'] = merged_code
        temp['Actual Total'] = merged_row['Counts']
        df_split_list.append(temp)

df_2021_disaggregated = pd.concat(df_split_list, ignore_index=True)

# Predict using model
X_test = df_2021_disaggregated[features]
df_2021_disaggregated['Predicted'] = model.predict(X_test)

# Normalize predictions within each merged district group
df_2021_disaggregated['Group Key'] = df_2021_disaggregated['Merged District Code']
group_sum = df_2021_disaggregated.groupby('Group Key')['Predicted'].transform('sum')
df_2021_disaggregated['Proportion'] = df_2021_disaggregated['Predicted'] / group_sum

# Final estimates using actual total
df_2021_disaggregated['Counts'] = (df_2021_disaggregated['Proportion'] * df_2021_disaggregated['Actual Total']).round().clip(lower=0).astype(int)

# Finalize 2021 dataset: drop merged entries and append split ones
df_2021_final = df2[~df2['Area Code'].isin(merged_codes)].copy()
df_2021_final = pd.concat([df_2021_final, df_2021_disaggregated[df2.columns]], ignore_index=True)

# Output the shape of final dataset
df_2021_final
```

```python
# Mapping from 2011 names to 2021 names (or vice versa)
name_mapping = {
    'Bristol, City of': 'Bristol',
    'Herefordshire, County of': 'Herefordshire',
    'Kingston upon Hull, City of': 'Kingston upon Hull',
    'Rhondda Cynon Taff': 'Rhondda Cynon Taf'
}
# Invert mapping if needed
reverse_mapping = {v: k for k, v in name_mapping.items()}
df_2021_final['Area Names'] = df_2021_final['Area Names'].replace(reverse_mapping)
```
